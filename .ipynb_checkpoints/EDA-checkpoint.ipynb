```python
from warnings import filterwarnings
filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
sns.set_context("poster")
sns.set_palette('Set2')
```

# import dataset

```python
df = pd.read_feather("dataset/carUsedPrice.feather")
df["model"] = df["model"].str.strip()
df = df[df["price"]<60000]
```

```python
df.head(4)
```

# visualize target label

```python
fig, ax = plt.subplots(figsize=(12, 6), constrained_layout=True)
_ = ax.set_title("Distribusion Plot Harga")
_ = sns.distplot(x=df["price"], bins=50, ax=ax);
_ = ax.set_xlabel("Harga")
_ = ax.set_ylabel("Jumlah Data")

plt.show()
```

# visualize kolom categori

```python
kolom = ["year", "mileage", "tax", "mpg", "engineSize"]

fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(16, 15), constrained_layout=True)
for kolom, ax in zip(kolom, ax.flatten()):
    sns.boxplot(x=df[df["price"]<75000][kolom], ax=ax);
    
plt.show()
```

# Hubungan Harga dan Transmisi

```python
fig, ax = plt.subplots(figsize=(12, 6), constrained_layout=True)

_ = fig.suptitle("Hubungan Harga dan Transmisi")
_ = sns.distplot(df[df["transmission"]=="Manual"]["price"], bins=40, label="Manual");
_ = sns.distplot(df[df["transmission"]=="Automatic"]["price"], bins=40, label="Matic");
_ = ax.set_xlabel("Harga")
_ = ax.set_ylabel("Jumlah Data")
_ = ax.legend()

plt.show()
```

# hubungan harga dan jenis mesin

```python
fig, ax = plt.subplots(figsize=(12, 6), constrained_layout=True)

_ = fig.suptitle("Hubungan Harga dan Jenis Mesin")
_ = sns.distplot(df[df["fuelType"]=="Petrol"]["price"], bins=40, label="Bensin");
_ = sns.distplot(df[df["fuelType"]=="Diesel"]["price"], bins=40, label="Diesel");
_ = sns.distplot(df[df["fuelType"]=="Hybrid"]["price"], bins=40, label="Hybrid");
_ = ax.set_xlabel("Harga")
_ = ax.set_ylabel("Jumlah Data")
_ = ax.legend()

plt.show()
```

# hapus outlier pada mpg

```python
Q1 = df["mpg"].quantile(0.25) #membuat Q1
Q3 = df["mpg"].quantile(0.75) #membuat Q3

IQR = Q3 - Q1 #menghitung IQR
print(f"IQR = {IQR}\nQ1 = {Q1}\nQ3 = {Q3}")

batas_bawah = Q1-(1.5*IQR)
batas_atas = Q3+(1.5*IQR)

print(f"batas bawah: {batas_bawah}\nbatas atas: {batas_atas}")
```

```python
df = df[(df["mpg"]>=batas_bawah) & (df["mpg"]<=batas_atas)] #filter outlier
df.reset_index(drop=True, inplace=True) #reset index
```

```python
Q1 = df["mileage"].quantile(0.25) #membuat Q1
Q3 = df["mileage"].quantile(0.75) #membuat Q3

IQR = Q3 - Q1 #menghitung IQR
print(f"IQR = {IQR}\nQ1 = {Q1}\nQ3 = {Q3}")

batas_bawah = Q1-(1.5*IQR)
batas_atas = Q3+(1.5*IQR)

print(f"batas bawah: {batas_bawah}\nbatas atas: {batas_atas}")

df = df[(df["mileage"]>=batas_bawah) & (df["mileage"]<=batas_atas)] #filter outlier
df.reset_index(drop=True, inplace=True) #reset index
```

```python
Q1 = df["year"].quantile(0.25) #membuat Q1
Q3 = df["year"].quantile(0.75) #membuat Q3

IQR = Q3 - Q1 #menghitung IQR
print(f"IQR = {IQR}\nQ1 = {Q1}\nQ3 = {Q3}")

batas_bawah = Q1-(1.5*IQR)
batas_atas = Q3+(1.5*IQR)

print(f"batas bawah: {batas_bawah}\nbatas atas: {batas_atas}")

df = df[(df["year"]>=batas_bawah) & (df["year"]<=batas_atas)] #filter outlier
df.reset_index(drop=True, inplace=True) #reset index

# =========================================

Q1 = df["tax"].quantile(0.25) #membuat Q1
Q3 = df["tax"].quantile(0.75) #membuat Q3

IQR = Q3 - Q1 #menghitung IQR
print(f"IQR = {IQR}\nQ1 = {Q1}\nQ3 = {Q3}")

batas_bawah = Q1-(1.5*IQR)
batas_atas = Q3+(1.5*IQR)

print(f"batas bawah: {batas_bawah}\nbatas atas: {batas_atas}")

df = df[(df["tax"]>=batas_bawah) & (df["tax"]<=batas_atas)] #filter outlier
df.reset_index(drop=True, inplace=True) #reset index
```

```python
kolom = ["year", "mileage", "tax", "mpg", "engineSize"]

fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(16, 15), constrained_layout=True)
for kolom, ax in zip(kolom, ax.flatten()):
    sns.boxplot(x=df[df["price"]<75000][kolom], ax=ax);
    
plt.show()
```
